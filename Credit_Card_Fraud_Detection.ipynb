```python


# Load the credit card fraud detection dataset
df = pd.read_csv('CC.csv')

# Convert the 'Time' column to hours
df['Time'] = df['Time'] / 3600

# Create a new DataFrame with the count of fraudulent transactions per hour
fraud_count = df[df['Class'] == 1].groupby('Time')['Class'].count().reset_index(name='Count')

# Plot the line chart
plt.figure(figsize=(18,7))
plt.plot(fraud_count['Time'], fraud_count['Count'])
plt.title('Number of Fraudulent Transactions over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Number of Fraudulent Transactions')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Load the credit card fraud detection dataset
df = pd.read_csv('CC.csv')

# Create a new DataFrame with the transaction amounts for fraudulent and non-fraudulent transactions
fraud_amounts = df.loc[df['Class'] == 1, 'Amount']
non_fraud_amounts = df.loc[df['Class'] == 0, 'Amount']
plt.figure(figsize=(15,5))
# Create a distribution plot with separate distributions for fraudulent and non-fraudulent transactions
sns.distplot(fraud_amounts, label='Fraudulent')
sns.distplot(non_fraud_amounts, label='Non-Fraudulent')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.legend()
plt.show()
```

This results in two new dataframes called fraud amounts and non fraud amounts.

## Feature transformations (Proposed)


### 1)	Grouping:

- We have around 28  transaction columns named  V1, V2...………...V28. By using the sum or mean functions we can group the columns , preferably according to the meaning of the feature. 

- For example, if our data contains lot of similar columns we can group two column, or group 3 columns to a single column to reduce the number of columns in our features.

- To increase the performance of the model we reduce the dimensionality

- Numerical Column Grouping


### 2)	Binning:
- Sorting transaction amounts into "low," "medium," and "high" categories depending on their values
- This feature engineering method involves categorizing transaction amounts into predetermined groups. For instance, transactions with quantities between $0  and  $100 can be classified as "low," those with amounts between $100 and $500 as "mid," and those with amounts over $500 as"high."
- Influence: The quantity of a transaction is a key indicator of credit card fraud, and binning it can help find patterns or anomalies that fall inside certain ranges and might point to fraudulent behavior.
- Binning is a type of feature engineering.

### 3)	Feature scaling:
- standardizing the data is necessary when we have data and features having high range.

- KNN,svm, logistic regression algorithms are sentive towards data , features with high range so scaling is important featue engineeing technique


### 4) Balancing:
- where ever the data is tend to be imabalnced , we are looking into some techniques which will eliminate the imabalnce problem. 
- one famous technique for balancing the dataset is SMOTE( Synthetic Minority Oversampling Technique)

```python

features['V1_V4'] = (features['V1']+features['V2']+features['V3']+features['V4'] )/4
features['V5_V8'] = (features['V5']+features['V6']+features['V7']+features['V8'] )/4
features['V9_V12'] = (features['V9']+features['V10']+features['V11']+features['V12'])/4
features['V13_V16'] = (features['V13']+features['V14']+features['V15']+features['V16'])/4
features['V17_V20'] = (features['V17']+features['V18']+features['V19']+features['V20'])/4
features['V21_V24'] = (features['V21']+features['V22']+features['V23']+features['V24'])/4
features['V25_V28'] = (features['V25']+features['V26']+features['V27']+features['V28'])/4
features.drop(['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9','V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',
               'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28'],axis = 1,inplace=True)
```

## 3) Feature scaling and SMOTE oversampling

### we splitted data into training and testing with 30% data goes to testing

```python

#doing Test Train.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=0)

from sklearn.preprocessing import StandardScaler
# enc = OneHotEncoder(handle_unknown='ignore')
scaler = StandardScaler()


labels_dict = {'Fraud':1,'Non-Fraud':0}

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)



print("instances per label in training set\n", y_train.value_counts())

print(X_train.shape)
print(X_test.shape)

print(labels_dict)

## training data sampling
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter

over = SMOTE(sampling_strategy='not majority', n_jobs=2)
steps = [('o', over)]
pipeline = Pipeline(steps=steps)
X_train, y_train = pipeline.fit_resample(X_train, y_train)
counter = Counter(y_train)
print(counter)

```

## Since all the required features scaling is done , hence we need apply alogrithms to the Training data nd also test the data


```python
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, plot_confusion_matrix, plot_roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
import statsmodels.api as sm
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
def run_model(model, X_train, y_train, X_test, y_test, verbose=True):
    t0=time.time()
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accuracy_train = model.score(X_train,y_train)
    accuracy_test = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred) 
    time_taken = time.time()-t0
    print("training Accuracy = {}".format(accuracy_train))
    print("testing Accuracy = {}".format(accuracy_test))
    print("ROC Area under Curve = {}".format(roc_auc))
    print("Time taken = {}".format(time_taken))
    print(classification_report(y_test,y_pred,digits=5))
    plot_confusion_matrix(model, X_test, y_test,cmap=plt.cm.pink)
    plot_roc_curve(model, X_test, y_test)                     
    
    return model, accuracy_test, roc_auc, time_taken
```

## logistic regression 

```python
model_lr = LogisticRegression()
model_lr, accuracy_lr, roc_auc_lr, tt_lr = run_model(model_lr, X_train, y_train, X_test, y_test)
```

## navie bayes

```python
model_nb = GaussianNB()
model_nb, accuracy_nb, roc_auc_nb, tt_nb = run_model(model_nb, X_train, y_train, X_test, y_test)
```

## K-Nearest Neighbor Classifier

```python
model_kn = KNeighborsClassifier()
model_kn, accuracy_kn, roc_auc_kn, tt_kn = run_model(model_kn, X_train, y_train, X_test, y_test)
```

## comparing accuracies of all models

```python

x = ['Logistic regression','Navie Bayes classifier',' KNearest Neighbour']
y =  [accuracy_lr*100,accuracy_nb*100,  accuracy_kn*100 ]
width = 0.75
plt.figure(figsize=(15,10))
fig, ax = plt.subplots(figsize=(15,7))

pps = ax.bar(x, y, width, align='center')

for p in pps:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+1,
      s="{}%".format(height),
      ha='center')
plt.title('accuraccies of all models')
plt.show()
```
